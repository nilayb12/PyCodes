```python
import os
import zipfile
import tkinter as tk

root = tk.Tk()
root.title("Python Compressed Files Extractor")
# root.geometry("600x400")

path_var = tk.StringVar()
extn_var = tk.StringVar()
extrPath_var = tk.StringVar()


def submit():
    path = path_var.get()
    extn = extn_var.get()
    extrPath = extrPath_var.get()
    path_var.set("")
    extn_var.set("")
    extrPath_var.set("")
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith(extn):
                zip_file_path = os.path.join(root, file)
                with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
                    zip_ref.extractall(extrPath)


path_label = tk.Label(
    root, text="Root Path of all Compressed Files: ", font=("Segoe UI Semibold", 10)
)
path_entry = tk.Entry(root, textvariable=path_var, font=("Segoe UI Semibold", 10))

extn_label = tk.Label(
    root, text="Compressed File Extension: ", font=("Segoe UI Semibold", 10)
)
extn_entry = tk.Entry(root, textvariable=extn_var, font=("Segoe UI Semibold", 10))

extrPath_label = tk.Label(
    root, text="Location to Store Extracted Data: ", font=("Segoe UI Semibold", 10)
)
extrPath_entry = tk.Entry(
    root, textvariable=extrPath_var, font=("Segoe UI Semibold", 10)
)

sub_btn = tk.Button(
    root, text="Extract Files", command=submit, font=("Segoe UI Semibold", 10)
)
exit_btn = tk.Button(
    root, text="Exit", command=root.destroy, font=("Segoe UI Semibold", 10)
)

path_label.grid(row=0, column=0)
path_entry.grid(row=0, column=1, pady=5)
extn_label.grid(row=1, column=0)
extn_entry.grid(row=1, column=1, pady=5)
extrPath_label.grid(row=2, column=0)
extrPath_entry.grid(row=2, column=1, pady=5)
sub_btn.grid(row=3, column=0, pady=5)
exit_btn.grid(row=3, column=1)

root.mainloop()
```
